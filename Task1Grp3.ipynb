```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import cartopy
import cartopy.crs as ccrs
import cartopy.feature as cf
import cartopy.io.shapereader as shapereader
from mpl_toolkits.axes_grid1 import make_axes_locatable

from pathlib import Path
```

```python
 data_rr=np.genfromtxt('rr_temp_rh_24_30May21.csv', delimiter=',')
```

```python
lat = data_rr[1: , 0]
lon = data_rr[1: , 1]
rr = data_rr[1:, 2]
tx = data_rr[1:, 3]
tn =data_rr[1:,4]
rh = data_rr[1:,5]
```

```python
#fig = plt.figure(figsize=(10,10))


#ax = plt.axes(projection=ccrs.PlateCarree())
fig, axes = plt.subplots(figsize = (20,8),ncols=4, subplot_kw = {'projection': ccrs.PlateCarree()})
#im = ax.imshow(datt)





# setting the lat and lon extent 
axes[0].set_extent([-3.6, 1.7, 4, 12], crs=ccrs.PlateCarree())
axes[1].set_extent([-3.6, 1.7, 4, 12], crs=ccrs.PlateCarree())
axes[2].set_extent([-3.6, 1.7, 4, 12], crs=ccrs.PlateCarree())
axes[3].set_extent([-3.6, 1.7, 4, 12], crs=ccrs.PlateCarree())


axes[0].add_feature(cf.BORDERS)
axes[1].add_feature(cf.BORDERS)
axes[2].add_feature(cf.BORDERS)
axes[3].add_feature(cf.BORDERS)

axes[0].add_feature(cf.COASTLINE)
axes[1].add_feature(cf.COASTLINE)
axes[2].add_feature(cf.COASTLINE)
axes[3].add_feature(cf.COASTLINE)

axes[0].set(title='  TotalRainfall  ')
axes[1].set(title='WMD Minimum temperature ')
axes[2].set(title='WMD Maximum temperature')
axes[3].set(title='mean RH')

#adding admin border from thr shapefile
#ax.add_geometries(adm1_shapes, ccrs.PlateCarree(), edgecolor='black', facecolor='none', alpha=0.8)


#adding the country boarder, the parameters given from cartopy


splot1 = axes[0].scatter(lon,lat,200,c=rr, cmap=cm.get_cmap('Blues'), alpha=1)
fig.colorbar( splot1,ax=axes[0], shrink=0.7, extend='both', label="rr")


splot2 = axes[1].scatter(lon,lat,200,c=tx, cmap=cm.get_cmap('Reds'), alpha=1)
fig.colorbar(splot2,ax=axes[1], shrink=0.7, extend='both', label="tx")                         
                         
                         
splot3 = axes[2].scatter(lon,lat,200,c=tn, cmap=cm.get_cmap('Reds'), alpha=1)
fig.colorbar(splot3,ax=axes[2], shrink=0.7, extend='both', label="tn")                        
                         

splot4 = axes[3].scatter(lon,lat,200,c=rh, cmap=cm.get_cmap('Blues'), alpha=1)
fig.colorbar(splot4,ax=axes[3], shrink=0.7, extend='both', label="RELATIVE HUMIDITY")





```

```python
fig, axes = plt.subplots(figsize = (10,10),nrows=2,ncols=2, subplot_kw = {'projection': ccrs.PlateCarree()})
#im = ax.imshow(datt)





# setting the lat and lon extent 
axes[0,0].set_extent([-3.6, 1.7, 4, 12], crs=ccrs.PlateCarree())
axes[0,1].set_extent([-3.6, 1.7, 4, 12], crs=ccrs.PlateCarree())
axes[1,0].set_extent([-3.6, 1.7, 4, 12], crs=ccrs.PlateCarree())
axes[1,1].set_extent([-3.6, 1.7, 4, 12], crs=ccrs.PlateCarree())


axes[0,0].add_feature(cf.BORDERS)
axes[0,1].add_feature(cf.BORDERS)
axes[1,0].add_feature(cf.BORDERS)
axes[1,1].add_feature(cf.BORDERS)

axes[0,0].add_feature(cf.COASTLINE)
axes[0,1].add_feature(cf.COASTLINE)
axes[1,0].add_feature(cf.COASTLINE)
axes[1,1].add_feature(cf.COASTLINE)

axes[0,0].set(title='  TotalRainfall  ')
axes[0,1].set(title='WMD Minimum temperature ')
axes[1,0].set(title='WMD Maximum temperature')
axes[1,1].set(title='mean RH')

#adding admin border from thr shapefile
#ax.add_geometries(adm1_shapes, ccrs.PlateCarree(), edgecolor='black', facecolor='none', alpha=0.8)


#adding the country boarder, the parameters given from cartopy


splot1 = axes[0,0].scatter(lon,lat,200,c=rr, cmap=cm.get_cmap('Blues'), alpha=1)
fig.colorbar( splot1,ax=axes[0,0], shrink=1, extend='both', label="RAINFALL(mm)")


splot2 = axes[0,1].scatter(lon,lat,200,c=tx, cmap=cm.get_cmap('Reds'), alpha=1)
fig.colorbar(splot2,ax=axes[0,1], shrink=1, extend='both', label="min TEMP")                         
                         
                         
splot3 = axes[1,0].scatter(lon,lat,200,c=tn, cmap=cm.get_cmap('Reds'), alpha=1)
fig.colorbar(splot3,ax=axes[1,0], shrink=1, extend='both', label="max TEMP")                        
                         

splot4 = axes[1,1].scatter(lon,lat,200,c=rh, cmap=cm.get_cmap('Blues'), alpha=1)
fig.colorbar(splot4,ax=axes[1,1], shrink=1, extend='both', label="RELATIVE HUMIDITY")
```
