```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python

Kumasi=pd.read_fwf("Kumasi_1983_2017_dRR_gapless.txt")#,names=['Year','Month','Day','Rainfall'])
```

```python
Kumasi=pd.read_fwf("Kumasi_1983_2017_dRR_gapless.txt",names=['Year','Month','Day','Rainfall'])
accra=pd.read_fwf("GAUGE_PRECIP/Accra_1983_2017_dRR_gapless.txt",names=['Year','Month','Day','Rainfall'])
Kete=pd.read_fwf("GAUGE_PRECIP/kete_krachi_1983_2017_dRR_gapless.txt",names=['Year','Month','Day','Rainfall'])
wa=pd.read_fwf("GAUGE_PRECIP/Wa_1983_2017_dRR_gapless.txt",names=['Year','Month','Day','Rainfall'])










```

```python
Kumasi["date"] = pd.to_datetime(Kumasi[["Year", "Month", "Day"]])
accra["date"] = pd.to_datetime(Kumasi[["Year", "Month", "Day"]])
Kete["date"] = pd.to_datetime(Kumasi[["Year", "Month", "Day"]])
wa["date"] = pd.to_datetime(Kumasi[["Year", "Month", "Day"]])
```

```python
df_k = Kumasi[[ "Rainfall", "date"]].set_index("date")
df_ac = accra[[ "Rainfall", "date"]].set_index("date")
df_ke = Kete[[ "Rainfall", "date"]].set_index("date")
df_wa = wa[[ "Rainfall", "date"]].set_index("date")
```

```python
df_k
```

```python
df_sum_k=df_k.sum(axis=1)
df_sum_ke=df_ke.sum(axis=1)
df_sum_ac=df_ac.sum(axis=1)
df_sum_wa=df_wa.sum(axis=1)
```

```python
df_k.sum(axis=1).plot(ylabel='anuall totall rainfall')
```

```python
df_sum_ac
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(df_sum_k, color="green", lw=1.5, ls="--", marker="*", label="KUMASI") #We specify some extra attributes here
ax.plot(df_sum_ac, color="red", lw=1.5, ls="--", marker="*", label="ACCRA")
ax.plot(df_sum_ke, color="blue", lw=1.5, ls="--", marker="*", label="KETE_KRACHI")
ax.plot(df_sum_wa, color="yellow", lw=1.5, ls="--", marker="*", label="WA")

# Label the axis
ax.set_ylabel("RAINFALL(mm)", fontsize=15)
ax.set_xlabel("Year", fontsize=15)
ax.set_title(
    "monthly RAINFALL TOTAL(MM) ", fontsize=15
)  # add title
ax.legend()  # add legend
#fig.savefig("Jan_Feb_Temp_Ano.pdf")
```

```python
df_sum_k1=df_k.resample('y').sum()
df_sum_ke1=df_ke.resample('y').sum()
df_sum_ac1=df_ac.resample('y').sum()
df_sum_wa1=df_wa.resample('y').sum()
```

```python
fig = plt.figure(figsize=(10, 5))

plt.plot(df_sum_k1, color="green", marker="*", label="KUMASI") #We specify some extra attributes here
plt.plot(df_sum_ac1, color="red", marker="^", label="ACCRA")
plt.plot(df_sum_ke1, color="blue", marker="p", label="KETE_KRACHI")
plt.plot(df_sum_wa1, color="yellow", marker="o", label="WA")

plt.legend()

# Label the axis
plt.ylabel("RAINFALL", fontsize=15)
plt.xlabel("Year", fontsize=15)
plt.title(
    "ANUALL RAINFALL TOTAL(MM) ", fontsize=15
)  # add title
```

```python
def ano (data):
    mean=np.mean(data)
    anomaly=data-mean
    return anomaly
```

```python
kum=ano(df_sum_k1)
ket=ano(df_sum_ke1)
acc=ano(df_sum_ac1)
waa=ano(df_sum_wa1)
```

```python
fig, axes = plt.subplots(figsize = (10,10),nrows=2,ncols=2, )


plt.subplots_adjust(right=1.5,hspace=0.5)

ax1=axes[0,0]
kum["color"] = np.where(kum["Rainfall"]<0, 'red', 'blue')
ax1.bar(kum.index, kum.Rainfall, color=kum.color, width=270)
ax1.set_xlabel("Year", fontsize=15)
ax1.set_ylabel("Rainfall Anomalies [mm]", fontsize=15)
ax1.set_title('Annual Rainfall Anomalies for Wa',fontsize=15)

ax2=axes[0,1]
ket["color"] = np.where(ket["Rainfall"]<0, 'red', 'blue')
ax2.bar(ket.index, ket.Rainfall, color=ket.color, width=270)
ax2.set_xlabel("Year", fontsize=15)
ax2.set_ylabel("Rainfall Anomalies [mm]", fontsize=15)
ax2.set_title('Annual Rainfall Anomalies for KETE_KRACHI',fontsize=15)

ax3=axes[1,0]
acc["color"] = np.where(acc["Rainfall"]<0, 'red', 'blue')
ax3.bar(acc.index, acc.Rainfall, color=acc.color, width=270)
ax3.set_xlabel("Year", fontsize=15)
ax3.set_ylabel("Rainfall Anomalies [mm]", fontsize=15)
ax3.set_title('Annual Rainfall Anomalies for ACCRA',fontsize=15)

ax4=axes[1,1]
waa["color"] = np.where(waa["Rainfall"]<0, 'red', 'blue')
ax4.bar(waa.index, waa.Rainfall, color=waa.color, width=270)
ax4.set_xlabel("Year", fontsize=15)
ax4.set_ylabel("Rainfall Anomalies [mm]", fontsize=15)
ax4.set_title('Annual Rainfall Anomalies for WA',fontsize=15)
















```
